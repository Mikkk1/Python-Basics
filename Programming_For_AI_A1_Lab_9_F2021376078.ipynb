# Reading AIML files And Response generation:

```python
from glob import glob
import aiml
k = aiml.Kernel()
```

```python
def read_files():
    path = "MyBot"
    for name in glob(path + "\*"):
        k.learn(name)
```

```python
def getResponse(query):
    while True:
        response = k.respond(query)
        if response:
            return (str(response))
        else:
            return (str(":)"))
```
