# Python Classes/Objects
Python is an object oriented programming language.

Almost everything in Python is an object, with its properties and methods.

A Class is like an object constructor, or a "blueprint" for creating objects.

```python
class MyClass:
  x = 5
```

```python
p1 = MyClass()
print(p1.x)
```

### **__init__** Function:

The ***self*** parameter is a reference to the current instance of the class, and is used to access variables that belong to the class.

```python
class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

  def __str__(self):
    return f"{self.name}({self.age})"

  # Object Methods:
  def my_name(self):
    print("Hello My name is " + self.name)




```

```python
# Creating new object, init function is automatically called, Like Constructor
p1 = Person('Sarim', 20)

del p1.age
p1.age = 21

p1.name, p1.age
```

```python
# Using the __str__ function we can print Obj as string:

print(p1)
```

```python
p1.my_name()
```

### Inheritence:

```python
class Person:
  def __init__(self, fname, lname):
    self.firstname = fname
    self.lastname = lname

  def printname(self):
    print(self.firstname, self.lastname)
```

```python
x = Person("Zahid", "Saeed")
x.printname()
```

```python
class inherited_class(Person):
  def __init__(self,fname,lname, grad):

    super().__init__(fname, lname)
    self.graduationyear = grad

  def welcome(self):
    print("Welcome", self.firstname, self.lastname, "to the class of", self.graduationyear)
```

```python
y = inherited_class("Sarim", "Zahid", 2025)
x.printname(), y.printname(), y.welcome()
```

# Iterator vs Iterable
Lists, tuples, dictionaries, and sets are all iterable objects. They are iterable containers which you can get an iterator from.

All these objects have a iter() method which is used to get an iterator:


```python
mytuple = ("Sarim", "Zahid", "Saeed")
myit = iter(mytuple)

print(next(myit))
print(next(myit))
print(next(myit))
```

```python
class MyNumbers:
  def __iter__(self):
    self.a = 1
    return self

  def __next__(self):
    if self.a <= 10:
      x = self.a
      self.a += 1
      return x
    else:
      raise StopIteration
```

```python
myclass = MyNumbers()
myiter = iter(myclass)

for x in myiter:
  print(x)
```

# Scope:
A variable created inside a function is available inside that function.

Global variables are available from within any scope, global and local.

The global keyword makes the variable global.

```python
x = 100
def myfunc():
  x = 300
  def innfunc():
    global x
    print(x)
  innfunc()
```

```python
myfunc()
```
