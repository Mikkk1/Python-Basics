### LOOPs

```python
i = 0
while i < 5:
  print('*')
  i += 1
```

```python
i = 0
while i < 5:
  print(i)
  if i == 3:
    break
  i += 1
```

```python
i = 0
while i < 5:
  i += 1
  if i == 3:
    continue
  print(i)
```

```python
l1 = ["Sarim", "Zahid", "Sidhu"]
for x in l1:
  print(x)
```

```python
for x in l1:
  if x == 'Zahid':
    break
  print(x)
```

```python
for x in l1:
  if x == 'Zahid':
    continue
  print(x)
```

# Functions:

```python
def print_pattern(x):
  for i in range(x):
    for j in range(i):
      print(i, end = ' ')
    print('\n')
```

```python
a = int(input("Enter row size: "))
print_pattern(a)
```

### Arbitrary Arguments *args:
If the number of arguments is unknown, add a * before the parameter name:

```python
def my_func(*args):
  print("The last thing i ate was " + args[2])
```

```python
my_func("Dahi Bhllay", "Nihaari", "Brownies")
```

### Keyword arguments:
You can also send arguments with the key = value syntax.

```python
def k_func(Iftaar, Dinner, Dessert):
  print("The dessert was: " + Dessert)
```

```python
k_func(Iftaar = "Dahi Bhllay", Dinner = "Nihaari", Dessert = "Brownies")
```

### Arbitrary Keyword Arguments, **kwargs
If you do not know how many keyword arguments that will be passed into your function, add two asterisk: ** before the parameter name in the function definition.

```python
def aka_func(**kargs):
  print("For iftar Buffet we chose " + kargs["p3"])
```

```python
aka_func(p1 = "Butt karahi", p2 = "Jade", p3 = "Villa The Grand")
```

# LAMBDA:

```python
def myfunc(n):
  return lambda a : a * n
```

```python
mydoubler = myfunc(2)
print(mydoubler(11))
```

```python
mytripler = myfunc(3)
mytripler(3)
```
