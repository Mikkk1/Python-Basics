#  Classical ML algorithms

Data Preparation

```python
train_X = [[2,1], [2,2], [2,3], [2,4]]
train_Y = [2, 4, 6, 8]
test_X = [[2, 8]]
test_Y = [16]
```

Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lg_model = LinearRegression()
```

```python
lg_model.fit(train_X, train_Y)
```

```python
lg_model.predict(test_X)[0]
```

```python
lg_model.score(train_X, train_Y)
```

Support Vector Machine

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC()
```

```python
svm_model.fit(train_X, train_Y)
```

```python
svm_model.predict(test_X)[0]
```

```python
svm_model.score(train_X, train_Y)
```

Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
```

```python
nb_model = GaussianNB()
```

```python
from sklearn.datasets import load_iris
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainX, trainY = load_iris(return_X_y=True)
```

```python
trainX
```

```python
trainY
```

```python
trainX, testX, trainY, testY = train_test_split(trainX, trainY, test_size=0.5, random_state=0)
```

```python
nb_model.fit(trainX, trainY)
```

```python
testX
```

```python
testY
```

```python
nb_model.predict(testX)
```

```python
nb_model.score(trainX, trainY)
```

Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier()
```

```python
dt_model.fit(trainX, trainY)
```

```python
y_Pred = dt_model.predict(testX)
```

```python
dt_model.score(trainX, trainY)
```

Un-Supervised (Clustering) -> Nearest Neighbours

```python
from sklearn.neighbors import NearestNeighbors
```

```python
import numpy as np
X = np.array([[-1, -1], [-2, -1], [-3, -2], [1, 1], [2, 1], [3, 2]])
```

```python
nn_model = NearestNeighbors(n_neighbors=2, algorithm='ball_tree')
```

```python
nn_model.fit(X)
```

```python
nn_model.kneighbors([[-2, -1]])
```
